```python
def iter_lines(txtdf):
    for stanza_i,stanzadf in txtdf.groupby('stanza_i'):
        lines=tqdm(list(stanzadf.groupby('line_i')),desc='Scanning through lines')
        for line_,linedf in lines:
            yield linedf
```

## Getting all stress possibilities

```python
def apply_combos(df,group1,group2,combo_key='combo_i'):
    # combo of indices?
    combo_opts = [
        [x for ii,x in grp.groupby(group2)]
        for i,grp in df.groupby(group1)
    ]
    
    # poss
    for combo in product(*combo_opts):
        yield pd.concat(combo)
```

```python
def iter_line_combos(df):
    for linedf in iter_lines(df):
        # get line combos
        for lci,dfcombo in enumerate(apply_combos(linedf,'word_i','word_ipa_i')):
            dfcombo['combo_i']=lci
            yield dfcombo
```

```python
for lc in iter_line_combos(txtdf): pass
```

### Getting windows

```python
def iter_windows(df,window_len=3):
    for dfslice in rolling_slices(df.reset_index(),window_len=window_len,keep_last_keys=['stanza_i','line_i','combo_i']):
        yield setindex(dfslice)
```

```python
for dfw in iter_windows(line_combo_df): pass
dfw
```

```python
def iter_possible_parses(df):
    for posi,posl in enumerate(possible_parses(len(df))):
        df['parse_i']=posi
        df['parse']=str(posl)
        df['syll_parse']=list(posl)
        df['is_w']=[np.int(x=='w') for x in posl]
        df['is_s']=[np.int(x=='s') for x in posl]
        yield setindex(df)
```
