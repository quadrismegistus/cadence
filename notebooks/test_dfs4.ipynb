```python
def iter_lines(txtdf):
    for stanza_i,stanzadf in txtdf.groupby('stanza_i'):
        lines=tqdm(list(stanzadf.groupby('line_i')),desc='Scanning through lines')
        for line_,linedf in lines:
            yield linedf
```

## Getting all stress possibilities

```python
def apply_combos(df,group1,group2,combo_key='combo_i'):
    # combo of indices?
    combo_opts = [
        [x for ii,x in grp.groupby(group2)]
        for i,grp in df.groupby(group1)
    ]
    
    # poss
    for combo in product(*combo_opts):
        yield pd.concat(combo)
```

```python
def iter_line_combos(df):
    for linedf in iter_lines(df):
        # get line combos
        for lci,dfcombo in enumerate(apply_combos(linedf,'word_i','word_ipa_i')):
            dfcombo['combo_i']=lci
            yield dfcombo
```

```python
line_combo_df = next(iter_line_combos(line_df))
line_combo_df
```

```python
def iter_possible_parses(phondf):
    for combodf in iter_line_combos(phondf):
        for posi,posl in enumerate(possible_parses(len(combodf))):
            combodf['parse_i']=posi
            combodf['parse']=str(posl)
            combodf['syll_parse']=list(posl)
            combodf['is_w']=[np.int(x=='w') for x in posl]
            combodf['is_s']=[np.int(x=='s') for x in posl]
            yield setindex(combodf)
```

```python
def iter_parsed(df,constraints=DEFAULT_CONSTRAINTS):
    for poss_parse in iter_possible_parses(df):
        yield apply_constraints(poss_parse, constraints=constraints)
```
