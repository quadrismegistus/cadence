```python
def parse_str(linepart_str,path=None,force=False,num_proc=1,engine=ENGINE,**kwargs):
    lpstr=linepart_str.strip()
    cs="|".join(sorted(list(kwargs.get("constraints",DEFAULT_CONSTRAINTS))))
    dbk=f'{lpstr}{DBSEP}parse{DBSEP}{cs}{DBSEP}{kwargs.get("engine")}'

    with get_db(path=path) as db:
        if force or not dbk in db:
            # odf
            linepartdf=get_scansion(lpstr,**kwargs)
            if engine==ENGINE_PROSODIC:
                odf=parse_prosodic_line(linepartdf,num_proc=num_proc,**kwargs)
            else:
                raise NotImplementedError('Only prosodic engine working at the moment...')
                o=list(iter_parsed_combos(linepartdf,num_proc=num_proc,progress=False,**kwargs))  
                odf=pd.concat(o) if len(o) else pd.DataFrame()
            db[dbk]=odf
        else:
            odf=db[dbk]
        return odf
```
